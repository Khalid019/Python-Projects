```python
print("Welcome to AB Bank\n\nInsert your card")

print("        Insert your pin number")
pin = int(input("         "))

password = 1234

choice = 0
balance = 500

if pin == password:

  while choice != 4:

    print("\n******Check the options*******")
    print(" 1 == Balance")
    print(" 2 == Deposit")
    print(" 3 == Withdraw")
    print(" 4 == Stop")

    choice = int(input("\nEnter your choice:\n "))
    if choice == 1:
      print("your current balance is: $",balance)
    elif choice == 2:
      deposit = int(input("Enter your deposit:\n "))
      balance += deposit
      print("\nyour current balance is: $",balance)  
    elif choice == 3:
      withdraw = int(input("Enter your withdraw:\n "))
      balance -= withdraw
      print("\nyour current balance is: $",balance) 
    else:
      print("Session ended. Thank you")   

else:
  print("Your password is incorrect")  
```
